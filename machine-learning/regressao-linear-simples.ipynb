#### 1 . Importações

```python
# Importando as biliotecas necessárias para a atividade.
import pandas as pd # A biblioteca pandas será usada para manipulação de dados
import numpy as np # A biblioteca numpy será usada para operações matemáticas
import matplotlib.pyplot as plt # A biblioteca matplotlib será usada para plotagem dos dados
from sklearn.linear_model import LinearRegression # A biblioteca sklearn será usada para criar e treinar o modelo
# Importando nossos dados meteorológicos
dados_meteorologicos = pd.read_csv("dados_meteorologicos.csv")

# Exibindo nossos dados (uma maneira de verificar se tudo ocorreu bem até agora)
#dados_meteorologicos
```

#### 2. Definindo os Eixos

Antes dos códigos, deixe-me deixar claro do que se tratará cada coisa:

* O eixo X (entrada) pode ser observado como sendo **as características e recursos**, ele se apresenta na horizontal e representa a **variável independente**.
* O eixo Y (saída) pode ser observado como sendo a **variável que estamos tentando prever**, trata-se da nossa **variável dependente**.
* Se fossemos fazer uma analogia simples, diriamos que **o valor de Y depende dos valores de X**.

```python
# O eixo x representará nossas entradas, para nosso exercício
# nós desejamos prever a umidade com base na temperatura
eixo_x = dados_meteorologicos['Temperatura(C)']

# O eixo y representará o valor a ser previsto, para nosso exercício
# desejamos descobrir qual será a umidade provável
eixo_y = dados_meteorologicos['Umidade']

# Plotando a umidade sobre a temperatura
plt.scatter(eixo_x,eixo_y)
# Adicionando rótulos para o gráfico
plt.xlabel('Temperatura (C)')
plt.ylabel('Umidade')
```

> Note que o gráfico de dispersão acima sugere uma tendência linear, vamos verificar essa relação

```python
# Utilizando o algoritmo de aprendizado de máquina
"""
 A linha de código abaixo pode parecer confusa, por isso explicarei:

 1. Estamos extraindo a coluna de temperatura do DataFrame
 2. Após isso, a coluna é convertida em um array NumPy
 3. Por fim, a forma do array é modificada (reshape) de modo
 que tenha uma coluna e quantas linhas forem necessárias

 OBS.: o uso de -1 em uma dimensão significa que o tamanho da dimensão
 também deve ser inferido para manter o número total de elementos
 
 O principal motivo por trás de tudo isso é que a maioria dos 
 modelos espera que a variável de entrada seja bidimensional
 por isso o uso do reshape(-1,1)
"""
x = dados_meteorologicos['Temperatura(C)'].values.reshape(-1,1)
y = dados_meteorologicos['Umidade']

modelo_regressao_linear = LinearRegression() # Instanciamos o objeto
modelo_regressao_linear.fit(x,y) # Ajustamos, isto é, treinamos o modelo
y_previsto = modelo_regressao_linear.predict(x)

# Plotando o gráfico com os dados reais e as previsões
plt.scatter(x, y, label='Dados Reais')
plt.plot(x, y_previsto, color='red', label='Regressão Linear')
plt.xlabel('Temperatura (C)')
plt.ylabel('Umidade')
plt.legend()
plt.show()
```

```python
# Suponha que você queira prever a umidade para uma temperatura específica, por exemplo, 32°C
temperatura_alvo = 32

# Usando o modelo treinado para prever a umidade
umidade_prevista = modelo_regressao_linear.predict([[temperatura_alvo]])  # Note que precisamos definir como um array numpy, por isso o colchete duplo

print(f'A umidade prevista para uma temperatura de {temperatura_alvo}°C é aproximadamente {umidade_prevista[0]:.2f}%')
```
