# 3. Observação com Análise Univariada

```python
_ = plt.hist(rh_dataset['Age'])
_ = plt.title("Histograma de Idade")
```

```python
_ = plt.hist(rh_dataset['MonthlyIncome'])
_ = plt.title("Histograma de Salário")
```

```python
rh_dataset['Gender'].value_counts()
```

```python
_ = rh_dataset['EducationField'].value_counts().plot.bar()
_ = plt.title('Headcount por Campo Educacional')
```

Com base nas verificações acima, vemos que a maioria dos funcionários é formado na área de Life Sciences.
Na sequência, temos muitos funcionários da área Medical e notamos que a área com menor quantia de funcionários formados é a de Human Resources.


# 4. Análise Bivariada

```python
_ = sns.boxplot(x='Gender', y='MonthlyIncome', data=rh_dataset)
```

Curiosamente, a distribuição salarial entre homens e mulheres é bem semelhante. Normalmente, espera-se encontrar salários mais altos para homens. Porém, no caso da IBM, o gender gap parece não ocorrer.
